```python
#Required Library Import
from bs4 import *
import requests as  rq
import os
```

```python
#Create variable for collecting website
DOMAIN = "https://www.heisig.com"
URL = "https://www.heisig.com/aktuelle-auftraege/drehen"

#Request the site and get text as html format
r2 = rq.get(URL)
soup2 = BeautifulSoup(r2.text,'html.parser')


links=[]

#Select the drawing image links by common words
x = soup2.select('a[href^="/Zeichnungen/"]')
for img in x:
    links.append(img['href'])
        
#Create local folder to store images    
os.mkdir('collected_image')

#Add domain link with image url to create actual download link
i=1
for l in links:
    #print(l)
    if l[:1]=="/":
        image=(DOMAIN + l)
        
    else:
        image=l      

    img_data = rq.get(image).content
    
#Store images with sequance integer number and jpg format        
    with open("collected_image\\"+str(i) +'.jpg', 'wb') as f:
        f.write(img_data)
    i=i+1
    

        
    

   


        
```
